```python
#UK100 data analysis from 2014 onwards
#Why am I doing this? Because there are 3 catastrophic events (2016 Brexit,2018 correction,2020 COVID)
```

```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = pd.to_datetime('2014-01-01')
end = pd.to_datetime('2020-06-01')
```

```python
df = pd.read_csv('UK100 Data.csv')
```

```python
df = df[(df['Date'] >= '2014-01-01' ) & (df['Date'] <= '2020-06-01')]
```

```python
df.head()
```

```python
df.tail()
```

```python
len(df)
```

```python
df.drop(columns=['Unnamed: 5','Unnamed: 6'],inplace=True)
```

```python
df.head()
```

```python
untradable_day = df[(df['Low'] > -120.1) | (df['High'] < 120.1)]
pct_crap = len(untradable_day)/len(df)
print(pct_crap)
```

```python
df['High'].quantile([0.2,0.5])
```

```python
df['Low'].quantile([0.8,0.5])
```

```python
bad_both_ways_days = df[(df['Low'] > -120.1) & (df['High'] < 120.1)]
bad_both = len(bad_both_ways_days)/len(df)
print(bad_both)
```
